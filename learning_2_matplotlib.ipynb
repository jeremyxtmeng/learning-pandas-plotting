### 1.1 intro

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

file_path = Path(r'C:\Users\xiang\Documents\backup_T580\Videos\Videos\Dropbox\p_jobinterview\coding\datacamp')

seattle_weather=pd.read_csv(file_path/'data'/'seattle_weather.csv')
seattle_weather.head(2)

```

```python
seattle_weather2=seattle_weather[seattle_weather['NAME']=='PALMER 3 ESE, WA US']
figure, ax=plt.subplots()

ax.plot(seattle_weather2['DATE'], seattle_weather2['MLY-TAVG-NORMAL'], 
        marker='v', 
        linestyle='--',
        color='r')
ax.set_xlabel('Month')
ax.set_ylabel('Degree')
ax.set_title('Seattle')
plt.show()
```

```python
seattle_weather3=seattle_weather[seattle_weather['NAME']=='EVERETT SNOHOMISH CO AIRPORT, WA US']

fig, ax = plt.subplots(2, 1,sharey=True)
ax[0].plot(seattle_weather2["DATE"], seattle_weather2["MLY-PRCP-NORMAL"],color='b')
ax[0].plot(seattle_weather2["DATE"], seattle_weather2["MLY-PRCP-25PCTL"], linestyle='--', color='b')
ax[0].plot(seattle_weather2["DATE"], seattle_weather2["MLY-PRCP-75PCTL"], linestyle='--', color='b')

ax[1].plot(seattle_weather3["DATE"], seattle_weather3["MLY-PRCP-NORMAL"], color='r')
ax[1].plot(seattle_weather3["DATE"], seattle_weather3["MLY-PRCP-25PCTL"], linestyle='--', color='r')
ax[1].plot(seattle_weather3["DATE"], seattle_weather3["MLY-PRCP-75PCTL"], linestyle='--', color='r')

ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
# plt.savefig('myfig.png')

```
