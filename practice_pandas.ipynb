# practice melt pivot_table

```python
import pandas as pd
import numpy as np

# Sample 1: Student test scores (wide format - good for melting)
students_wide = pd.DataFrame({
    'student_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'math_score': [85, 92, 78, 95, 88],
    'english_score': [90, 85, 88, 92, 95],
    'science_score': [88, 90, 85, 89, 92],
    'history_score': [92, 88, 90, 94, 87]
})

print("Original Wide Format:")
print(students_wide)
```

```python
df1=students_wide.melt(id_vars='student_id',
                   value_vars=['math_score', 'english_score', 'science_score', 'history_score'],
                   var_name='subject',
                   value_name='score')
print(df1)
```

```python
df1.pivot_table(values='score', index='student_id', columns='subject', )
```

```python
# Sample 2: Sales data across quarters
sales_wide = pd.DataFrame({
    'product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Accessories'],
    'Q1_2024': [100, 250, 150, 80, 300],
    'Q2_2024': [120, 280, 140, 90, 320],
    'Q3_2024': [110, 300, 160, 85, 310],
    'Q4_2024': [130, 320, 170, 95, 330]
})

print("\nSales Data (Wide Format):")
print(sales_wide)

df1=sales_wide.melt(id_vars=['product', 'category'],
            value_vars=['Q1_2024', 'Q2_2024','Q3_2024','Q4_2024'],
            var_name='season',
                value_name='quantity')

print(df1.pivot_table(values='quantity',index=['product', 'category'], columns='season')
)
```

```python
# Sample 3: Temperature data across cities
weather_wide = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=7, freq='D'),
    'New_York': [32, 35, 30, 28, 33, 36, 34],
    'Los_Angeles': [65, 68, 70, 67, 69, 71, 68],
    'Chicago': [25, 28, 22, 20, 27, 30, 26],
    'Miami': [75, 78, 80, 77, 79, 82, 76]
})

weather_wide.melt(id_vars='date', 
                  value_vars=['New_York', 'Los_Angeles', 'Chicago', 'Miami'],
                  value_name='value',
                  var_name='city'
                  )\
    .pivot_table(values='value',index='date', columns='city')
```

# practice rolling windows and cleaning

```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Create base stock data
stock_data = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=10, freq='D'),
    'open': np.random.randint(90, 110, 10),
    'close': np.random.randint(90, 110, 10),
    'volume': np.random.randint(1000, 5000, 10)
})

# 1. Introduce missing values (NaN)
stock_data.loc[2, 'open'] = np.nan
stock_data.loc[5, 'close'] = np.nan
stock_data.loc[7, 'volume'] = np.nan
stock_data.loc[8, ['open', 'close']] = np.nan

# 2. Add duplicate rows
duplicate_row = stock_data.iloc[3].copy()
stock_data = pd.concat([stock_data, pd.DataFrame([duplicate_row])], ignore_index=True)
duplicate_row2 = stock_data.iloc[6].copy()
stock_data = pd.concat([stock_data, pd.DataFrame([duplicate_row2])], ignore_index=True)

# 3. Add extreme outlier values
stock_data.loc[1, 'volume'] = 50000  # Extremely high volume
stock_data.loc[4, 'open'] = 5        # Unrealistically low price
stock_data.loc[9, 'close'] = 500     # Extremely high price

# 4. Add some additional data quality issues
stock_data.loc[len(stock_data)] = [pd.Timestamp('2024-01-05'), 95, 98, -100]  # Negative volume
stock_data.loc[len(stock_data)] = [pd.NaT, 100, 105, 2000]  # Missing date

# Shuffle the data to make it more realistic
stock_data = stock_data.sample(frac=1).reset_index(drop=True)
```

```python
# unique values
stock_data.drop_duplicates(subset=['date'],inplace=True)
print(stock_data['date'].nunique())
print(len(stock_data['date']))
print(stock_data['date'].isna().sum())
```

```python
stock_data['date'].astype(str).replace('NaT', np.nan).fillna('missing value')
stock_data['date'].astype(str).replace('NaT', np.nan).dropna()
```

```python
stock_data2=stock_data[(~stock_data['date'].isna()) 
                       & (~stock_data['open'].isna())
                       & (~stock_data['close'].isna()) ]
stock_data2.info()
stock_data2.head()
stock_data2=stock_data2.sort_values(by='date').reset_index(drop=True)
```

```python
median_open=stock_data2['open'].median()
median_close=stock_data2['close'].median()

open_low=stock_data2['open'].min()
open_high=stock_data2['open'].quantile(0.9)

close_low=stock_data2['close'].quantile(0.1)
close_high=stock_data2['close'].quantile(0.9)
# clean data to replace extreme value with median
condition_list=[stock_data2['open']<=open_low, 
                stock_data2['open']>open_high]
choice_list=[median_open,median_open]

condition_list2=[stock_data2['close']<close_low, 
                stock_data2['close']>close_high]
choice_list2=[median_close,median_close]

stock_data2['open2']=np.select(condition_list, choice_list,default=stock_data2['open'])
stock_data2['close2']=np.select(condition_list2, choice_list2,default=stock_data2['close'])

stock_data2.loc[:,'open2':'close2'].describe()
```

```python

# Daily returns
stock_data2['daily_return'] =(stock_data2['close2']-stock_data2['open2'])/stock_data2['open2']

# Intraday change
stock_data2['intraday_change'] = stock_data2['open2'].diff(1)

# 5-day moving average
stock_data2['momentum_5d'] =stock_data2['open2'].rolling(5).mean()

stock_data2
```
