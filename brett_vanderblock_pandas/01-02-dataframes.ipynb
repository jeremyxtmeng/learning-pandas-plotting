
## <center> Intro to Dataframes

```python
import pandas as pd
```

```python
scores = {"name":['Ray','Japhy','Zosa'],
          "city":['San Francisco','San Francisco','Denver'],
          "score":[75,92]
         }
```

```python
try:
    df = pd.DataFrame(scores)
except Exception as e:
    print("Error: array must the same length")
    scores['score'].append(30)
    df = pd.DataFrame(scores)
```

note: converting to df from a arrays of same length

```python
df
```

```python
# access columns
df['score']
```

```python
# create a new column
df['name_city'] = df['name'] + '_' + df['city']
```

```python
# return rows when a condition is met
df[df['score']>90]
```

```python
df[df['score']>90]['score']
```
