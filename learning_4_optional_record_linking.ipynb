#### (optional) record linking

````python
import pandas as pd
from pathlib import Path 
from thefuzz import process # Import process from thefuzz
import numpy

file_name=Path.cwd()
restaurants=pd.read_csv(file_name/'data'/'restaurants_L2_dirty.csv')
restaurants_new=pd.read_csv(file_name/'data'/'restaurants_L2.csv')

# clean cuisine names
# Store the unique values of cuisine_type in unique_types
unique_types = restaurants['type'].unique()
print(len(unique_types))

categories=['asian', 'american','italian']

matches= process.extract(categories[0], restaurants['type'], limit = len(restaurants['type']))

print(matches[0:3]) # (closest match, similarity score, index of match)

for category in categories:
    # Calculate similarity of 'asian' to all values of unique_types
    matches= process.extract(category, restaurants['type'], limit = len(restaurants['type']))

    for match in matches:
        if match[1]>=80:
            restaurants.loc[restaurants['type'] == match[0],'type'] = category

# Inspect the final result
print(len(restaurants['type'].unique()))
````
